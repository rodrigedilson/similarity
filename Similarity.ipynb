# Cálculo de Similaridade

```python
import numpy as np
```

***Gerando os Embeddings***

```python
def calc_embedding(text, max):
  encoded_input = tokenizer(text, return_tensors='pt', padding='max_length', max_length = max)
  # Gera os embeddings de palavras
  output = model(**encoded_input)
  # O último estado oculto é o primeiro elemento do output, isso tem o embedding para cada token
  embedding = output.last_hidden_state
  emb_numpy = embedding.cpu().detach().numpy()
  return emb_numpy
```

```python
embeddings = []
for d in docs:
  embeddings.append(calc_embedding(d, 512))
```

```python
embeddings[:1]
```

***Cálculo da Similaridade de Cosseno***

```python
from scipy.spatial.distance import cosine

def cosine_similarity2(vec1, vec2):
    # Computa a similaridade do cosseno entre dois vetores
    return 1 - cosine(vec1, vec2)
```

***Analisando a similaridade nos parágrafos***

```python
for i in range(0, len(embeddings) - 2, 2):
  #flatten é para deixar todos na mesma dimensão
  vec1 = embeddings[i].flatten()
  vec2 = embeddings[i+2].flatten()
  similar = cosine_similarity2(vec1, vec2)
  print(f'Páragrafos {i + 1} e {i + 3}:')
  print(docs[i])
  print('-------------------------')
  print(docs[i+2])
  print('--------------------------')
  print(f'Similaridade: {similar}')
  print('-------------------------\nfim')
```

# Tópicos com LSI

***Tokenizando no formato LSI***

```python
texts = [
    [word for word in d.lower().split()]
    for d in doc_limpo
]
```

```python
texts[:5]
```

***Cálculo de frequência das palavras***

```python
frequency = defaultdict(int)
for text in texts:
    for token in text:
        frequency[token] += 1
```

***Removendo Hapax Legomena***

```python
texts = [
    [token for token in text if frequency[token] > 1]
    for text in texts
]
```

```python
texts[:3]
```

***Criando o Corpus***

```python
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

***Criando o corpus Tf–idf***

```python
tfidf = models.TfidfModel(corpus)  # step 1 -- initialize a model
```

```python
corpus_tfidf = tfidf[corpus]
```

***Extraindo e exibindo os tópicos com LSI***





```python
lsi_model = models.LsiModel(corpus_tfidf, id2word=dictionary, num_topics=10)  # initialize an LSI transformation
corpus_lsi = lsi_model[corpus_tfidf]  # create a double wrapper over the original corpus: bow->tfidf->fold-in-lsi
```

```python
top = lsi_model.print_topics(10)
top
```

***Extraindo os tópicos mais relevantes***

```python
topicos = []
for t in top:
  topic_str = t[1]
  parts = topic_str.split('+')
  keywords = [part.split('*')[1].replace('"','') for part in parts]
  topicos.append(keywords)
topicos
```

```python
respostas = []
for q in questions:
  response = openai.ChatCompletion.create(
    model='gpt-3.5-turbo-16k',
      messages = [
    {'role': 'user', 'content': q}
  ],
    temperature=0.5
  )
  response_text = response['choices'][0]['message']['content']
  respostas.append(response_text)
respostas
```
